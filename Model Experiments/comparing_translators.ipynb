```python
import pandas as pd
from transformers import (
    pipeline,
    AutoTokenizer,
    AutoModelWithLMHead,
    TranslationPipeline,
    FastSpeech2ConformerHifiGan,
)
import csv
```

```python
def add_text_to_csv(text, csv_file_path):
    # Open the CSV file for appending
    with open(csv_file_path, mode='a', newline='', encoding='utf-8') as csv_file:
        writer = csv.writer(csv_file)
        # Write the text to the CSV
        writer.writerow([text])
```

```python
asr_text = pd.read_csv('asr_output.csv', header=None, names=['ASR_Transcription'])
asr_text = asr_text.iloc[1:]
asr_text.reset_index(inplace=True)
asr_text.drop('index',inplace=True, axis=1)
asr_text
```

```python
opus_translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-es", device='mps')
t5_translator = pipeline("translation", model="vgaraujov/t5-base-translation-en-es", device='mps', max_length=400)
bart_translator = pipeline("translation", model="DrishtiSharma/mbart-large-50-en-es-translation-lr-1e-05-weight-decay-0.0", device='mps', max_length=400)
```

```python
def translate_1(text):
    spanish_text = opus_translator(text)
    return spanish_text

def translate_2(text):
    spanish_text = t5_translator(text)
    return spanish_text
```

```python
print('Opus Translator...')
for i, row in asr_text.iterrows():
    print(f'Translating text {i}')
    translated_text = translate_1(row['ASR_Transcription'])
    add_text_to_csv(translated_text, 'translations/opus_translations.csv')
```

```python
print('T5 Translator...')
for i, row in asr_text.iterrows():
    print(f'Translating text {i}')
    translated_text = translate_2(row['ASR_Transcription'])
    add_text_to_csv(translated_text, 'translations/t5_translations.csv')
```
